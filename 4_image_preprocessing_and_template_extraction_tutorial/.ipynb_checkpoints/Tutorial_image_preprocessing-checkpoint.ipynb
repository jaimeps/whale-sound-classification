The goal of this notebook is to explain the techniques used (by the winner of the Kaggle competition) to pre-process the images (spectrograms) and increase the classification accuracy.

## PART 1: Focus on the low frequency spectrum

Two functions pre-process the images: one vertically and one horizontally. We'll go through the techniques used step by step with one example. 

The first thing that is noteworthy is that the winner focuses on the lower range of the frequencies of the spectrogram, since the whale calls are commonly located in this area of the plot. 

```python
# Set path to directory with folders train and test
path_data = '/Users/jaime/Documents/datasets/mlprojectdata/data/'
```

```python
import aifc
import matplotlib.pyplot as plt
import matplotlib.ticker as plticker
from matplotlib import mlab, cm
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
# ReadAIFF function
def ReadAIFF(file):
# Reads the frames from the audio clip and returns the uncompressed data
    s = aifc.open(file,'r')
    nFrames = s.getnframes()
    strSig = s.readframes(nFrames)
    return np.fromstring(strSig, np.short).byteswap()

# Read one file as an example
params = {'NFFT':256, 'Fs':2000, 'noverlap':192} 
s = ReadAIFF(path_data + 'train/train6.aiff')
P, freqs, bins = mlab.specgram(s, **params)
```

Each spectrogram from the dataset returns a 129x59 matrix

```python
Q = P.copy()
m, n = Q.shape
print 'Shape: ', m, n
```

The winner focuses on the mean and standard deviation of the lower range of the image. Let's see what this means graphically

```python
# Mean and std of specific ranges of the image are extracted
maxM = 40
mval, sval = np.mean(Q[:maxM,:]), np.std(Q[:maxM,:])

# Function to allow code reusability
def plot_spectrogram(ax, P):
    plt.imshow(P, origin='lower', extent=[-6,6,-1,1], aspect=4, cmap = cm.get_cmap('bwr'))
    loc = plticker.MultipleLocator(base=3.0) # this locator puts ticks at regular intervals
    ax.xaxis.set_major_locator(loc)
    ax.set_xticklabels(np.arange(-0.5,2.5,0.5))
    ax.set_yticklabels(range(0,1001,250))
    ax.set_xlabel('Time (seconds)', fontsize = 12)
    ax.set_ylabel('Frequency (Hz)', fontsize = 12)
    cbar = plt.colorbar()
    cbar.set_label('Amplitude', fontsize = 12)

# Let's see what this means
R = Q.copy()
R[maxM:,:] = 2500
fig = plt.figure(figsize = (14,4))
ax1 = plt.subplot(121)
plot_spectrogram(ax1, R)
ax2 = plt.subplot(122)
plot_spectrogram(ax2, P)
plt.suptitle('Comparison low frequency range (left) vs full image (right)', fontsize = 16)
plt.show()
```

## PART 2: Contrast enhancement

Since the amplitude of the spectrogram is a relative measure, extreme values make the spectrogram lose detail (the image only highlights those extreme values).

To enhance the contrast and appreciate more detail in the spectrogram, the winner submission caps the value in the spectrogram to   1.5  

This makes a huge difference in the aspect of the spectrogram.

```python
# Extreme values are capped to mean  1.5 std
fact_ = 1.5
Q[Q > mval + fact_*sval] = mval + fact_*sval
Q[Q < mval - fact_*sval] = mval - fact_*sval

# Let's see it
R = Q.copy()
R[maxM:,:] = 2500
fig = plt.figure(figsize = (14,4))
ax1 = plt.subplot(121)
plot_spectrogram(ax1, P)
ax2 = plt.subplot(122)
plot_spectrogram(ax2, Q)
plt.suptitle('Comparison orignal image (left) vs capped extreme values (right)', fontsize = 16)
plt.show()
```

As we can see in the image above, capping the extreme values allows us to extract subtle details in the spectrogram.

## PART 3: Introducing numpy's convolve function

The code makes use of numpy's convolve function. This returns the discrete, linear convolution of two one-dimensional sequences. Basically, it slides one vector over the other, adding the multiplication of the elements.

Let's see it a simple example.

```python
# Set two vectors
z = [4, 3]
a = [1, 1, 5, 5]

# Compare
print np.convolve(a, z)

# Flip one vector
z2 = [3, 4]

# Slide flipped vector through a, adding the multiplication of the elements
print z2[0] * 0 + z2[1] * a[0]
print z2[0] * a[0] + z2[1] * a[1]
print z2[0] * a[1] + z2[1] * a[2]
print z2[0] * a[2] + z2[1] * a[3]
print z2[0] * a[3] + z2[1] * 0
```

```python
# If we use 'same' option, we remove the non-overlapping
print np.convolve(a, z, 'same')
```

Interesting behavior when we use a vector of ones: we obtain at once a sum of the elements with a sliding window

```python
z = [1, 1]
a = [2, 3, 4, 5, 6]
# Compare
print np.convolve(a, z, 'same')
print [a[0], a[0] + a[1], a[1] + a[2], a[2] + a[3], a[3] + a[4]]
```

Using the parameters used in the winner's code:

```python
# Matrix with ones
inner=3
outer=64
wInner = np.ones(inner)
wOuter = np.ones(outer)

# Convolve with a matrix of ones just produces a sum with rolling window
i = 0
print 'First element'
print '   ', np.convolve(Q[:,i],wInner,'same')[0]
print '   ', sum(Q[0:2,i])
print 'Second element'
print '   ', np.convolve(Q[:,i],wInner,'same')[1]
print '   ', sum(Q[0:3,i])
print 'Third element'
print '   ', np.convolve(Q[:,i],wInner,'same')[2]
print '   ', sum(Q[1:4,i])
print 'Fourth element'
print '   ', np.convolve(Q[:,i],wInner,'same')[3]
print '   ', sum(Q[2:5,i])
```

## PART 4: Removing local mean

By using two windows, a large and a small one, we can subtract the local mean of each point. The purpose here is to increase the differences within the shape of the whale call in the spectrogram.

```python
# The difference between sums of different windows (64 and 3) is subtracted from each element
R = Q.copy()
for i in range(n):
    Q[:,i] = Q[:,i] - (np.convolve(R[:,i],wOuter,'same') - np.convolve(R[:,i],wInner,'same'))/(outer - inner)
# Let's see it
fig = plt.figure(figsize = (14,4))
ax1 = plt.subplot(121)
plot_spectrogram(ax1, R)
ax2 = plt.subplot(122)
plot_spectrogram(ax2, Q)
plt.suptitle('Comparison before (left) and after (right) removing local mean from the image ', fontsize = 16)
plt.show()
```

```python
# Let's compare original with final
fig = plt.figure(figsize = (14,4))
ax1 = plt.subplot(121)
plot_spectrogram(ax1, P)
ax2 = plt.subplot(122)
plot_spectrogram(ax1, Q)
plt.suptitle('Comparison original (left) and after (right) vertical treatment ', fontsize = 16)
plt.show()
```

Finally, the lower frequency of this modified spectrogram is passed to the next step: template matching.

```python
# The lower part of the image is returned then
P2 = P.copy()
Q2 = Q.copy()
P2[maxM:, :] = 2500
Q2[maxM:, :] = 160
fig = plt.figure(figsize = (14,4))
ax1 = plt.subplot(121)
plot_spectrogram(ax1, P2)
ax2 = plt.subplot(122)
plot_spectrogram(ax2, Q2)
plt.suptitle('Comparison original (left) and after (right) output low frequency ', fontsize = 16)
plt.show()
```

Then the process is repeated horizontally, producing a second image per each spectrogram

```python
# Repeat the process horizontally
inner=3
outer=32
maxM=60
PH, freqs, bins = mlab.specgram(s, **params)
QH = PH.copy()
m, n = Q.shape

mval, sval = np.mean(QH[:maxM,:]), np.std(QH[:maxM,:])
fact_ = 1.5
QH[QH > mval + fact_*sval] = mval + fact_*sval
QH[QH < mval - fact_*sval] = mval - fact_*sval
wInner = np.ones(inner)
wOuter = np.ones(outer)
for i in range(maxM):
    QH[i,:] = QH[i,:] - (np.convolve(QH[i,:],wOuter,'same') - np.convolve(QH[i,:],wInner,'same'))/(outer - inner)

# Let's compare original with final
fig = plt.figure(figsize = (14,4))
ax1 = plt.subplot(121)
plot_spectrogram(ax1, PH)
ax2 = plt.subplot(122)
plot_spectrogram(ax2, QH)
plt.suptitle('Comparison original (left) and after (right) horizontal treatment ', fontsize = 16)
plt.show()    
```

The window to be returned has slightly different dimensions but also belongs to the low frequency spectrum.

```python
# Compare original vs output
PH2 = PH.copy()
QH2 = QH.copy()
PH2[maxM:, :] = 2500
QH2[maxM:, :] = 90
fig = plt.figure(figsize = (14,4))
ax1 = plt.subplot(121)
plot_spectrogram(ax1, PH2)
ax2 = plt.subplot(122)
plot_spectrogram(ax2, QH2)
plt.suptitle('Comparison original (left) and after (right) output low frequency ', fontsize = 16)
plt.show()
```

Finally, let's plot the two original images, coupled with the two final outputs

```python
fig = plt.figure(figsize = (14,4))
ax1 = plt.subplot(121)
plot_spectrogram(ax1, P)
ax2 = plt.subplot(122)
plot_spectrogram(ax2, Q2)
plt.suptitle('Comparison original image (left) and output (right) after vertical treatment', fontsize = 16)
plt.show()

fig = plt.figure(figsize = (14,4))
ax1 = plt.subplot(121)
plot_spectrogram(ax1, PH)
ax2 = plt.subplot(122)
plot_spectrogram(ax2, QH2)
plt.suptitle('Comparison original image (left) and output (right) after horizontal treatment', fontsize = 16)
plt.show()
```
